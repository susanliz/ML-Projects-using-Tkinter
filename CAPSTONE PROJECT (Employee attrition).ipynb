### Data Preprocessing

```python
df['Attrition'].unique()
```

```python
def yesorno(a):
    if a == 'Yes':
        return 1
    else:
        return 0
    
df['Attrition'] = df['Attrition'].apply(yesorno)
```

```python
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
newdf=pd.get_dummies(df,drop_first=True)
newdf.head()
```

#### Check the shape of our new dataset

```python
newdf.shape
```

#### Print unique values in our new dataframe

```python
for i in df.columns:
    print(f"Unique values in {i} are {df[i].unique()}")
```

#### Split the columns into input and target variables (X and y)

```python
x = newdf.drop(columns=['Attrition'])
y = newdf['Attrition']
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
X =std.fit_transform(x)
```

```python
X = pd.DataFrame(X,columns=x.columns)
X.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2)
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
y_pred = model.predict(xtest)
```

```python
print(f"Accuracy of our model is {model.score(xtrain,ytrain)} \nConfusion matrix is {confusion_matrix(ytest,y_pred)} \nClassification Report is {classification_report(ytest,y_pred)} \nCross Validation score is {cross_val_score(model,x,y,cv=5).mean()}")
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
sns.heatmap(confusion_matrix(ytest,y_pred),annot=True,fmt='d')
```
